```python
# import the sales file
import pandas as pd
sales = pd.read_csv("/content/Walmart_Sales.csv")
sales.head()
```

```python
# Convert Date to datetime object and format
sales['Date'] = pd.to_datetime(sales['Date'], format='%m-%d-%Y').dt.strftime('%m-%d-%Y')

# Round Weekly_Sales, Fuel_Price to 2 decimal places
sales['Weekly_Sales'] = sales['Weekly_Sales'].round(2)
sales['Fuel_Price'] = sales['Fuel_Price'].round(2)

# Round Temperature to the nearest whole number
sales['Temperature'] = sales['Temperature'].round(0).astype(int)

# Round CPI, Unemployment to 3 decimal places
sales['CPI'] = sales['CPI'].round(3)
sales['Unemployment'] = sales['Unemployment'].round(3)

# Display the updated DataFrame
sales
```

```python
# Assuming 'Holiday_Flag' column indicates holiday weeks (1 for holiday, 0 for non-holiday)
sales['Holiday_Flag'] = ...  # Fill in logic to identify holiday weeks

# Group by holiday flag and calculate average weekly sales
holiday_impact = sales.groupby('Holiday_Flag')['Weekly_Sales'].mean()

# Create a bar chart
import matplotlib.pyplot as plt
holiday_impact.plot(kind='bar')
plt.xlabel('Holiday Week')
plt.ylabel('Average Weekly Sales')
plt.title('Impact of Holidays on Weekly Sales')
plt.show()
```

```python
# Group by store and calculate average unemployment rate
store_unemployment = sales.groupby('Store')['Unemployment'].mean()

# Display the unemployment rate for each store
store_unemployment
```

```python
# Find the store with the lowest unemployment rate
lowest_unemployment_store = store_unemployment.idxmin()
lowest_unemployment_rate = store_unemployment.min()

print("Store with lowest unemployment rate:", lowest_unemployment_store, "(", lowest_unemployment_rate, ")")
```

```python
# Find the store with the highest unemployment rate
highest_unemployment_store = store_unemployment.idxmax()
highest_unemployment_rate = store_unemployment.max()

print("Store with highest unemployment rate:", highest_unemployment_store, "(", highest_unemployment_rate, ")")
```

```python
print("Store with lowest unemployment rate:", lowest_unemployment_store, "(", lowest_unemployment_rate, ")")
print("Store with highest unemployment rate:", highest_unemployment_store, "(", highest_unemployment_rate, ")")
```

```python
overall_correlation = sales['CPI'].corr(sales['Weekly_Sales'])
print("Overall correlation between CPI and Weekly Sales:", overall_correlation)
```

```python
non_holiday_sales = sales[sales['Holiday_Flag'] == 0]
non_holiday_correlation = non_holiday_sales['CPI'].corr(non_holiday_sales['Weekly_Sales'])
print("Correlation between CPI and Weekly Sales (Non-Holiday):", non_holiday_correlation)
```

```python
# Calculate correlation matrix
correlation_matrix = sales[['Fuel_Price', 'Weekly_Sales', 'CPI', 'Temperature', 'Unemployment']].corr()

# Display correlation matrix
print(correlation_matrix)
```

```python
import statsmodels.formula.api as sm

# Define the regression model
model = sm.ols('Weekly_Sales ~ Fuel_Price + CPI + Temperature + Unemployment', data=sales)

# Fit the model
results = model.fit()

# Print the summary of the regression results
print(results.summary())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create scatter plot with regression line
sns.lmplot(x='Fuel_Price', y='Weekly_Sales', data=sales, hue='Holiday_Flag')
plt.xlabel('Fuel Price')
plt.ylabel('Weekly Sales')
plt.title('Fuel Price vs Weekly Sales')
plt.show()
```
